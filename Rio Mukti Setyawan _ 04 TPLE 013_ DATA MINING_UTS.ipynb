### Soal
Dari data di atas lakukan pre-processing sbb:
1. Lakukan analisa mengenai field mana saja yang memiliki data kosong(Nan) lalu Isi data yang kosong tersebut menggunakan nilai mean
2. lakukan oneHotEncoder ke field state
3. buat field baru yaitu Tax. dimana Tax = (profit + Marketing Spend + administration) * 5%
4. Jalankan standardScaler field-field tersebut

````python
# Jawab
# 1
# Import library
import pandas as pd
import numpy as np

# Load data
data_startup = pd.read_csv('50_Startups.csv')


# Identifikasi kolom-kolom numerik
numeric_columns = data_startup.select_dtypes(include=[np.number]).columns

# Isi data kosong dengan nilai mean untuk kolom-kolom numerik
data_startup[numeric_columns] = data_startup[numeric_columns].fillna(data_startup[numeric_columns].mean())

````

````python
# One-Hot Encoder
data_startup = pd.get_dummies(data_startup, columns=['State'])
````

````python
# Buat kolom baru 'Tax'
data_startup['Tax'] = (data_startup['Profit'] + data_startup['Marketing Spend'] + data_startup['Administration']) * 0.05
````

````python
from sklearn.preprocessing import StandardScaler

# Inisialisasi StandardScaler
scaler = StandardScaler()

# StandardScaler untuk field-field tersebut
scaled_features = scaler.fit_transform(data_startup[['R&D Spend', 'Administration', 'Marketing Spend', 'Profit', 'Tax']])

# Update DataFrame dengan data yang telah di-scale
data_startup[['R&D Spend', 'Administration', 'Marketing Spend', 'Profit', 'Tax']] = scaled_features
````
