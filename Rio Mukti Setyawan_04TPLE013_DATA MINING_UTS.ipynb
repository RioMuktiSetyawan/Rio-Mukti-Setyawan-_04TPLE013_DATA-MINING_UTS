### Soal
Dari data di atas lakukan pre-processing sbb:
1. Lakukan analisa mengenai field mana saja yang memiliki data kosong(Nan) lalu Isi data yang kosong tersebut menggunakan nilai mean
2. lakukan oneHotEncoder ke field state
3. buat field baru yaitu Tax. dimana Tax = (profit + Marketing Spend + administration) * 5%
4. Jalankan standardScaler field-field tersebut

````python
#Menjawab pertanyaan Nomor 1
import pandas as pd

# Load the data
data_startup = pd.read_csv('50_Startups.csv')

# Analyze missing data
print("Missing data in each column:")
print(data_startup.isna().sum())

# Fill NaN values with the mean of the column for numeric columns only
numeric_columns = data_startup.select_dtypes(include=['number']).columns
data_startup[numeric_columns] = data_startup[numeric_columns].fillna(data_startup[numeric_columns].mean())

# Verify missing values are filled
print("\nData after filling missing values:")
print(data_startup.isna().sum())
````

````python
#Menjawab pertanyaan Nomor 2
from sklearn.preprocessing import OneHotEncoder

# OneHotEncode the 'State' field
one_hot_encoder = OneHotEncoder(sparse_output=False)
state_encoded = one_hot_encoder.fit_transform(data_startup[['State']])

# Create a DataFrame with the encoded state columns
state_encoded_df = pd.DataFrame(state_encoded, columns=one_hot_encoder.get_feature_names_out(['State']))

# Drop the original 'State' column and concatenate the new encoded columns
data_startup = data_startup.drop('State', axis=1)
data_startup = pd.concat([data_startup, state_encoded_df], axis=1)

# Display the data to check the new state columns
print("\nData with OneHotEncoded 'State' column:")
print(data_startup.head())
````

````python
#Menjawab pertanyaan Nomor 3
# Create a new field 'Tax'
data_startup['Tax'] = (data_startup['Profit'] + data_startup['Marketing Spend'] + data_startup['Administration']) * 0.05

# Display the data to check the new 'Tax' column
print("\nData with new 'Tax' column:")
print(data_startup.head())

````

````python
#Menjawab pertanyaan Nomor 4
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Select the numeric fields to be standardized
numeric_fields = ['R&D Spend', 'Administration', 'Marketing Spend', 'Profit', 'Tax']

# Standardize the numeric fields
data_startup[numeric_fields] = scaler.fit_transform(data_startup[numeric_fields])

# Display the processed data
print("\nProcessed data after standardization:")
print(data_startup.head())

````
